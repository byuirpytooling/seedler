```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import random as r
```

```python
seed = r.randint(0, 100_000)
r.seed(seed)
```

```python
test_count = 10_000_000
numbers = [r.randint(1, 100) for _ in range(test_count)]
```

```python
series = pd.Series(numbers)
counts = series.value_counts().sort_index()
```

```python
counts.plot(kind='bar')

plt.title(f"Distribution of Random Numbers (Seed={seed})")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.xticks(range(0, 100, 5))

plt.show()
```

```python
std = series.std()
print("Standard Deviation: %s" % std) # expected near 28.87
print("(%s)" % (std / (test_count / 100)))
```

```python
mean = series.mean()
print(f"Mean: {mean}")
```
